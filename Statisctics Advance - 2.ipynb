### Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.

### Ans:- 
* PMF: In probability theory, a probability mass function (PMF) describes the probability of obtaining a particular discrete value in a discrete random variable. It gives the probability of each outcome.

##### Example: Consider a fair six-sided die. The PMF of rolling the die and getting a 3 is 1/6.

* PDF: A probability density function (PDF) is used for continuous random variables. It describes the likelihood of a variable falling within a particular range. Unlike PMF, PDF doesn't give the probability at specific points but rather the probability density over a range.

##### Example: The PDF of a standard normal distribution is the bell-shaped curve associated with it.


### Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?


### Ans:- 
* The Cumulative Density Function (CDF) gives the probability that a random variable takes a value less than or equal to a given point. It provides a cumulative probability distribution.

##### Example: For a six-sided die, the CDF at 3 is the probability of rolling a 1, 2, or 3.

#### CDF is used to calculate probabilities for ranges of values and is crucial in statistics and hypothesis testing.

### Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli Distribution and Binomial Distribution?


### Ans:- The Bernoulli distribution is a discrete probability distribution representing a binary outcome, where an experiment results in either success (with probability p) or failure (with probability q = 1 - p). It is named after Jacob Bernoulli, a Swiss mathematician.

#### Example: Consider a single coin toss, where getting heads (H) is considered a success, and tails (T) is a failure. Let:

* Success (H) with probability p = 0.5 (fair coin).
* Failure (T) with probability q = 1 - p = 0.5.


##### Difference between Bernoulli Distribution and Binomial Distribution:

The Bernoulli distribution represents a single trial (e.g., one coin toss).
The Binomial distribution represents the number of successes in a fixed number of independent Bernoulli trials.

It is a continuous distribution, and its probability density function (PDF) is constant over the entire range.

### Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.


### Ans:-   The Central Limit Theorem (CLT) is a fundamental concept in probability and statistics. It states that the distribution of the sum (or average) of a large number of independent, identically distributed random variables approaches a normal distribution, regardless of the original distribution of the individual variables.

1. Normality Approximation: Regardless of the original distribution of the population, the distribution of the sample mean tends to follow a normal distribution as the sample size increases. This allows statisticians to make probabilistic statements about sample means using the properties of the normal distribution.

2. Statistical Inference: The CLT is the foundation for many statistical techniques and hypothesis tests. It enables the use of normal distribution-based methods even when dealing with non-normally distributed populations, as long as the sample size is sufficiently large.

3. Population Inference: The CLT is crucial in making inferences about population parameters based on sample statistics. It facilitates the estimation of confidence intervals and hypothesis testing for means and proportions.

4. Random Sampling Assurance: The CLT provides assurance that, under certain conditions, the distribution of sample means is well-behaved and less dependent on the shape of the original population distribution. This is especially valuable in practical applications where the underlying distribution may be unknown or complex.

#### In summary, the Central Limit Theorem is a powerful tool that allows statisticians to make reliable inferences about populations, even when the population distribution is not known or is not normal, by relying on the normal distribution of sample means for sufficiently large samples.


### Q10: State the assumptions of the Central Limit Theorem.

### Ans:-
1. Independence: The random variables must be independent.

2. Sample Size: The larger the sample size, the better the approximation to normality.

3. Population Distribution: The shape of the original population distribution doesn't matter if the sample size is large enough.
